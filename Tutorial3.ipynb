```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
from io import StringIO
import itertools
def find_titanic_csv():
    candidates = [
        Path.cwd(),
        Path.cwd() / "sample_data",
        Path("/content/sample_data"),
        Path("/mnt/data"),
    ]
    names = ["titanic.csv", "Titanic.csv"]
    # 1) First check likely directories directly
    for d, n in itertools.product(candidates, names):
        p = d / n
        if p.exists():
            return p
    # 2) Then search recursively inside those directories
    for root in candidates:
        if root.exists():
            hits = list(root.rglob("titanic*.csv"))
            if hits:
                return hits[0]
    # 3) We search recursively from current working directory
    for hit in Path.cwd().rglob("titanic*.csv"):
        return hit
    raise FileNotFoundError(
        "Could not find 'titanic.csv'. Put it in the working dir, sample_data/, "
        "/content/sample_data, /mnt/data, or provide an absolute path."
    )
DATA_PATH = find_titanic_csv()
print(f"Using file: {DATA_PATH}")
df = pd.read_csv(DATA_PATH)
# Understand the Data
print("\n=== Preview (head) ===")
print(df.head(10))
print("\n=== Data Types & Null Counts (.info) ===")
buf = StringIO()
df.info(buf=buf)
print(buf.getvalue())
print("=== Missing Values: count and percentage ===")
missing = df.isna().sum().to_frame("missing_count")
missing["missing_pct"] = (missing["missing_count"] / len(df) * 100).round(2)
print(missing.sort_values("missing_pct", ascending=False))
#We impute missing data: numeric -> median, categorical -> mode
df_imputed = df.copy()
num_cols = df_imputed.select_dtypes(include=[np.number]).columns
cat_cols = df_imputed.select_dtypes(include=["object", "category"]).columns
df_imputed[num_cols] = df_imputed[num_cols].fillna(df_imputed[num_cols].median())
for c in cat_cols:
    df_imputed[c] = df_imputed[c].fillna(df_imputed[c].mode().iloc[0])
# Summary Statistics
print("\n=== Descriptive Statistics (numeric) ===")
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
print(df[num_cols].describe().T)
print("\n=== Categorical Value Counts (top 10 per column) ===")
cat_cols = df.select_dtypes(include=["object", "category"]).columns.tolist()
for likely_cat in ["Survived", "Pclass", "SibSp", "Parch"]:
    if likely_cat in df.columns and likely_cat not in cat_cols:
        cat_cols.append(likely_cat)
for c in cat_cols:
    print(f"\n-- {c} --")
    print(df[c].value_counts(dropna=False).head(10))
# Visualizations (matplotlib)
# Univariate — numeric histograms
for col in num_cols:
    plt.figure()
    df[col].hist(bins=30)
    plt.title(f"Histogram — {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.tight_layout()
    plt.show()
# Univariate — boxplots
for col in [c for c in ["Age", "Fare"] if c in df.columns]:
    plt.figure()
    df[[col]].boxplot(column=col)
    plt.title(f"Boxplot — {col}")
    plt.ylabel(col)
    plt.tight_layout()
    plt.show()
# Univariate — categorical bars
for col in [c for c in ["Survived", "Sex", "Embarked", "Pclass"] if c in df.columns]:
    ax = df[col].value_counts(dropna=False).plot(kind="bar")
    ax.set_title(f"Bar Chart — {col}")
    ax.set_xlabel(col)
    ax.set_ylabel("Count")
    plt.tight_layout()
    plt.show()
# Bivariate Analysis
# Survived vs Sex
if set(["Survived", "Sex"]).issubset(df.columns):
    ax = pd.crosstab(df["Sex"], df["Survived"]).plot(kind="bar")
    ax.set_title("Survived vs Sex — counts")
    ax.set_xlabel("Sex")
    ax.set_ylabel("Count")
    plt.tight_layout()
    plt.show()
# Survived vs Pclass
if set(["Survived", "Pclass"]).issubset(df.columns):
    ax = pd.crosstab(df["Pclass"], df["Survived"]).plot(kind="bar")
    ax.set_title("Survived vs Pclass — counts")
    ax.set_xlabel("Pclass")
    ax.set_ylabel("Count")
    plt.tight_layout()
    plt.show()
# Age vs Fare scatter
if set(["Age", "Fare"]).issubset(df.columns):
    plt.figure()
    if "Survived" in df.columns:
        alive = df[df["Survived"] == 1]
        dead  = df[df["Survived"] == 0]
        plt.scatter(alive["Age"], alive["Fare"], alpha=0.6, label="Survived=1")
        plt.scatter(dead["Age"],  dead["Fare"],  alpha=0.6, label="Survived=0")
        plt.legend()
    else:
        plt.scatter(df["Age"], df["Fare"], alpha=0.6)
    plt.title("Age vs Fare — scatter")
    plt.xlabel("Age")
    plt.ylabel("Fare")
    plt.tight_layout()
    plt.show()
# Stacked share by category
def stacked_share(feature):
    if "Survived" not in df.columns:
        return
    ct = pd.crosstab(df[feature], df["Survived"], normalize="index")
    ax = ct.plot(kind="bar", stacked=True)
    ax.set_title(f"Survival share by {feature}")
    ax.set_xlabel(feature)
    ax.set_ylabel("Share")
    plt.tight_layout()
    plt.show()

for feat in [f for f in ["Sex", "Pclass", "Embarked"] if f in df.columns]:
    stacked_share(feat)
# Correlation Matrix (numeric)
if len(num_cols) >= 2:
    corr = df[num_cols].corr()
    print("\n=== Correlation Matrix (numeric) ===")
    print(corr)

    plt.figure()
    im = plt.imshow(corr, aspect="auto")
    plt.title("Correlation Heatmap (numeric features)")
    plt.xticks(range(len(corr.columns)), corr.columns, rotation=45, ha="right")
    plt.yticks(range(len(corr.index)), corr.index)
    plt.colorbar(im, fraction=0.046, pad=0.04)
    plt.tight_layout()
    plt.show()
# Outliers (IQR rule)
def outlier_summary(series: pd.Series):
    s = series.dropna()
    if s.empty:
        return {"Q1": np.nan, "Q3": np.nan, "IQR": np.nan,
                "LowerFence": np.nan, "UpperFence": np.nan, "OutliersCount": 0}
    q1, q3 = np.percentile(s, [25, 75])
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    outliers = ((s < lower) | (s > upper)).sum()
    return {"Q1": q1, "Q3": q3, "IQR": iqr,
            "LowerFence": lower, "UpperFence": upper, "OutliersCount": int(outliers)}

check_cols = [c for c in ["Age", "Fare", "SibSp", "Parch"] if c in df.columns]
rows = [{"feature": c, **outlier_summary(df[c])} for c in check_cols]
outlier_df = pd.DataFrame(rows)
print("\n=== Outlier Summary (IQR rule) ===")
print(outlier_df)

print("\nDone.")
```
